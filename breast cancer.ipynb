```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
# Replace with the correct dataset filename
df = pd.read_csv("C:/Users/Bhanu Mutha/OneDrive/Documents/breast cancer.csv")
df.head()
```

```python
# Remove ID or unnamed columns if present
df.drop(columns=[col for col in df.columns if 'id' in col.lower() or 'unnamed' in col.lower()], inplace=True)
```

```python
# Replace '?' with NaN and drop rows with missing values
df.replace('?', np.nan, inplace=True)
df.dropna(inplace=True)
```

```python
# Convert diagnosis column to numeric values: M = 1 (Malignant), B = 0 (Benign)
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
k = 5  # You can change this and test different values
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
